```python
import pandas as pd
import pickle

combined_sent = pd.read_pickle('Combined_wrt_topics.pkl')
sentences = []
for i in range(len(combined_sent)):
    sentences.append(combined_sent.transcript.loc[i])
sentences
```

```python
#original

from scipy import spatial
from sent2vec.vectorizer import Vectorizer

vectorizer = Vectorizer()
vectorizer.run(sentences)
vectors_bert = vectorizer.vectors
for i in range(len(combined_sent)-1):
    dist_1 = spatial.distance.cosine(vectors_bert[i], vectors_bert[i+1])
    print('dist_{0}-{1} : {2}'.format(i+1, i+2, dist_1))
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
sentence_embeddings = model.encode(sentences)
```

```python
bert_data = []
bert_data.append(combined_sent.loc[0].at['transcript'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
a = 0
sentences_string = ""
sentences_string += combined_sent.loc[0].at['transcript']
final = []
b = 0

for i in range(1, len(combined_sent)):
    sentence_embeddings2 = model.encode(bert_data)
    print('cs {} - {}'.format(i, i+1))
    print(cosine_similarity([sentence_embeddings[i]],sentence_embeddings2[a:a+1])[0][0])
    print('value of a is {}'.format(a+1))
    print(sentences_string)
    
    if(b != a):
        final[a] = sentences_string
        b = a
    else: 
        final.append(sentences_string)
        
    if(cosine_similarity([sentence_embeddings[i]],sentence_embeddings2[a:a+1])[0][0] >= 0.7):
        sentences_string += combined_sent.loc[i].at['transcript']
        bert_data[a] = sentences_string
        
    else:
        a += 1
        sentences_string = combined_sent.loc[i].at['transcript']
        bert_data.append(sentences_string)
    b += 1

if(b != a):
    final[a] = sentences_string
    b = a
    
else: 
    final.append(sentences_string)
```

```python
# bert_data
data = {}
for i in range(len(final)):
    data[i] = final[i]
data
```

```python
# We are going to change this to key: sentence_id, value: string format
def combine_text(list_of_text):
    '''Takes a list of text and combines them into one large chunk of text.'''
    combined_text = ''.join(list_of_text)
    return combined_text
```

```python
data_combined = {key: [combine_text(value)] for (key, value) in data.items()}
data_combined
```

```python
# We can either keep it in dictionary format or put it into a pandas dataframe
import pandas as pd
pd.set_option('max_colwidth',150)

combined_sent_bert = pd.DataFrame.from_dict(data_combined).transpose()
combined_sent_bert.columns = ['transcript']
combined_sent_bert = combined_sent_bert.sort_index()
combined_sent_bert
```

```python
# pickle it for later use
combined_sent_bert.to_pickle("Combined_wrt_cosine_similarity_bert.pkl")
```
